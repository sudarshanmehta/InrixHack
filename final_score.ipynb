```python
import pandas as pd

data = pd.read_csv('consumer_combined_data.csv')

data['end_date_time'] = pd.to_datetime(data['end_date_time'])
```

```python
reference_date = data['end_date_time'].max()
data['recency_days'] = (reference_date - data['end_date_time']).dt.days
```

```python
recency_score_min = 1.0
recency_score_max = 10.0
data['recency_score'] = recency_score_max - ((data['recency_days'] - data['recency_days'].min()) / 
                                             (data['recency_days'].max() - data['recency_days'].min()) * 
                                             (recency_score_max - recency_score_min))
```

```python
import numpy as np

# Example: Setting lambda to a value (you might need to adjust this)
lambda_ = 0.05

# Applying exponential decay
data['decay_factor'] = np.exp(-lambda_ * data['recency_days'])

# Adjusting decayed score to fit in the 1.0 to 10.0 range
data['decayed_recency_score'] = recency_score_max - ((1 - data['decay_factor']) * (recency_score_max - recency_score_min))
```

```python
# Counting the number of visits for each device_id to each restaurant_id
visit_counts = data.groupby([ 'restaurant_id']).size().reset_index(name='visit_count')

# # Normalizing the visit_count to a scale of 1 to 10
# frequency_score_min = 1.0
# frequency_score_max = 10.0
# visit_counts['frequency_score'] = frequency_score_max - ((visit_counts['visit_count'] - visit_counts['visit_count'].min()) / 
#                                                          (visit_counts['visit_count'].max() - visit_counts['visit_count'].min()) * 
#                                                          (frequency_score_max - frequency_score_min))

# # Merging this back with the main data if needed
# data = data.merge(visit_counts[['device_id', 'restaurant_id', 'frequency_score']], on=['device_id', 'restaurant_id'])
```

```python
max_distance_threshold = data['trip_distance_meters'].quantile(0.95)  # 95th percentile as threshold

# Cap the distance at the threshold
data['capped_distance'] = data['trip_distance_meters'].clip(upper=max_distance_threshold)

# Normalize the capped distance to a scale of 1 to 10
distance_score_min = 1.0
distance_score_max = 10.0
data['distance_score'] = ((data['capped_distance'] - data['capped_distance'].min()) / 
                          (data['capped_distance'].max() - data['capped_distance'].min()) * 
                          (distance_score_max - distance_score_min)) + distance_score_min

# Scores at or above the threshold will be set to 10.0
data.loc[data['trip_distance_meters'] >= max_distance_threshold, 'distance_score'] = 10.0
```

```python
data.head(5)
```

```python
data['traffic_score'] = (data['trip_mean_speedkph'] / data['trip_mean_speedkph'].max() ) * 10 # Normalizing to a 0-1 scale
```

```python
data.head(5)
```

```python
# Redefining weights for each feature
weights = {
    'distance': 0.20,
    'recency': 0.40,
    # 'frequency': 0.05,
    'traffic_score': 0.40
}
```

```python
# Calculating the composite score
data['composite_score'] = (data['distance_score'] * weights['distance'] +
                                       data['decayed_recency_score'] * weights['recency'] +
                                    #    data['frequency_score'] * weights['frequency'] +
                                       data['traffic_score'] * weights['traffic_score'])
```

```python
data = data.sort_values('composite_score', ascending=False)
```
